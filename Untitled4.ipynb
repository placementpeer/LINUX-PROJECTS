```python
# Step 1: Install & import
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files

# Step 2: Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]   # gets the uploaded file's name

# Step 3: Load image
img = cv2.imread(image_path)

# Step 4: Basic operations
resized = cv2.resize(img, (300, 300))                             # Resize
cropped = img[50:200, 100:300]                                    # Crop [y1:y2, x1:x2]

(h, w) = img.shape[:2]
M = cv2.getRotationMatrix2D((w // 2, h // 2), 45, 1.0)             # Rotate 45°
rotated = cv2.warpAffine(img, M, (w, h))

# Step 5: Display results
print("Original:")
cv2_imshow(img)

print("Resized:")
cv2_imshow(resized)

print("Cropped:")
cv2_imshow(cropped)

print("Rotated:")
cv2_imshow(rotated)
```
